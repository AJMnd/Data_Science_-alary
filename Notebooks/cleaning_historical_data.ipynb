```python
# Import our input dataset
hist_wages_df = pd.read_csv('../Resources/historicalWages.csv')
hist_wages_df.head()
```

```python
#drop the lonely 2009 row
hist_wages_df = hist_wages_df[1:]
hist_wages_df.head()
```

```python
#drop columns
hist_wages_df = hist_wages_df.drop(columns=['Quarter',
    'Number of workers (in thousands) - Men',
    'Number of workers (in thousands) - Women',
    'Median weekly earnings (in current dollars) - Men',
    'Median weekly earnings (in current dollars) - Women',
    'Median weekly earnings (in constant dollars) - Men',
    'Median weekly earnings (in constant dollars) - Women'])
hist_wages_df.head()
```

```python
#rename columns
hist_wages_df = hist_wages_df.rename(columns = {
    "Number of workers (in thousands) - Total": "Total # of Workers",
"Median weekly earnings (in current dollars) - Total": "Total Median Weekly Earnings (in current dollars)",
"Median weekly earnings (in constant dollars) - Total": "Total Median Weekly Earnings (in constant dollars)"})
hist_wages_df.head()
```

```python
#want annual salary
hist_wages_df['Total Median Annually Earnings (in current dollars)'] = hist_wages_df['Total Median Weekly Earnings (in current dollars)']*52

hist_wages_df['Total Median Annually Earnings (in constant dollars)'] = hist_wages_df['Total Median Weekly Earnings (in constant dollars)']*52

hist_wages_df.head()
```

```python
per_year_wages = hist_wages_df.groupby(['Year']).mean()
per_year_wages
```
