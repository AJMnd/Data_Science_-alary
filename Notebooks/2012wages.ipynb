```python
import pandas as pd
import numpy as np
```

```python
file_path = "../newdata/oesm12st/state_M2012_dl.xls"
wages12_df = pd.read_excel(file_path)
wages12_df
```

```python
# Cleaning DataFrame by dropping unnecessary columns df.drop(columns=['B', 'C'], axis=1) axis=1 drops the whole column Stackoverflow

wages12_df = wages12_df.drop(columns=['AREA', 'ST', 'OCC_CODE', 'OCC_GROUP', 'EMP_PRSE', 'JOBS_1000', 'H_PCT75', 'H_PCT90', 'A_PCT10', 'A_PCT25', 'A_PCT75',  'A_PCT90', 'ANNUAL', 'HOURLY'], axis=1)
wages12_df
```

```python
wages12_df = wages12_df.drop(columns=['LOC_Q', 'MEAN_PRSE', 'H_PCT10', 'H_PCT25'], axis=1)
wages12_df
```

```python
wages12_df
```

```python
wages12_df.info()
```

```python
# Getting rid of duplicates, Pandas documentation https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop_duplicates.html

wages12_df = wages12_df.drop_duplicates()
```

```python
# Resetting index after dropping duplicates. https://stackoverflow.com/questions/28885073/reindexing-after-pandas-drop-duplicates
wages12_df = wages12_df.reset_index(drop=True)
```

```python
wages12_df
```

```python
# 
wages12_df = wages12_df.dropna(axis=0, how='any')
wages12_df.head()
```

```python
# Rename the headers
wages12_df = wages12_df.rename(columns={"STATE" : "State",
"OCC_TITLE": "Occupation",
"TOT_EMP": "Total Employees",
"H_MEAN": "Hourly Mean",
"A_MEAN": "Annual Mean",
"H_MEDIAN": "Hourly Median",
"A_MEDIAN": "Annual Median",})

wages12_df.head()
```

```python
# Replacing special characters with nan and then dropping them.
wages12_df = (wages12_df
 .replace('#', np.nan)
 .replace('*', np.nan)
 .replace('**', np.nan)
              
 .dropna()
)

wages12_df.head()
```

```python
# Resetting the index after dropping rows.
wages12_df = wages12_df.reset_index(drop=True)
```

```python
# Renaming the columns  
wages12_df.astype({
    'Hourly Mean':float,
    'Hourly Median':float,
    'Total Employees':float,
    'Annual Mean':float,
    'Annual Median':float,
}).dtypes
```

```python
# Exporting Data Frame to csv 
wages12_df.to_csv("../Clean Data/2012wages.csv", index = False)
```
