```python
import pandas as pd
import numpy as np
```

```python
file_path = "../newdata/oesm19st/state_M2019_dl.xlsx"
wages19_df = pd.read_excel(file_path)
wages19_df
```

```python
# Cleaning DataFrame by dropping unnecessary columns df.drop(columns=['B', 'C'], axis=1) axis=1 drops the whole column Stackoverflow

wages19_df = wages19_df.drop(columns=['area', 'area_type', 'naics', 'naics_title', 'i_group', 'own_code', 'occ_code', 'o_group', 'h_pct75', 'h_pct90', 'a_pct10', 'a_pct25', 'a_pct75'], axis=1)
```

```python
wages19_df
```

```python
wages19_df = wages19_df.drop(columns=['emp_prse', 'jobs_1000', 'loc_quotient', 'pct_total', 'mean_prse', 'h_pct10', 'h_pct25', 'a_pct90', 'annual', 'hourly'], axis=1)
wages19_df
```

```python
wages19_df.info()
```

```python
# Getting rid of duplicates, Pandas documentation https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop_duplicates.html

wages19_df = wages19_df.drop_duplicates()
```

```python
# Resetting index after dropping duplicates. https://stackoverflow.com/questions/28885073/reindexing-after-pandas-drop-duplicates
wages19_df = wages19_df.reset_index(drop=True)
```

```python
wages19_df
```

```python
# 
wages19_df = wages19_df.dropna(axis=0, how='any')
wages19_df.head()
```

```python
# Rename the headers
wages19_df = wages19_df.rename(columns={"area_title" : "State",
"occ_title": "Occupation",
"tot_emp": "Total Employees",
"h_mean": "Hourly Mean",
"a_mean": "Annual Mean",
"h_median": "Hourly Median",
"a_median": "Annual Median",})

wages19_df.head()
```

```python
# Replacing special characters with nan and then dropping them.
wages19_df = (wages19_df
 .replace('#', np.nan)
 .replace('*', np.nan)
 .replace('**', np.nan)
              
 .dropna()
)

wages19_df.head()
```

```python
# Resetting the index after dropping rows.
wages19_df = wages19_df.reset_index(drop=True)
```

```python
# Renaming the columns  
wages19_df.astype({
    'Hourly Mean':float,
    'Hourly Median':float,
    'Total Employees':float,
    'Annual Mean':float,
    'Annual Median':float,
}).dtypes
```

```python
wages19_df
```

```python
# Exporting Data Frame to csv 
wages19_df.to_csv("../Clean Data/2019wages.csv", index = False)
```
