```python
import pandas as pd
import numpy as np
```

```python
file_path = "../newdata/oesm21st/state_M2021_dl.xlsx"
wages21_df = pd.read_excel(file_path)
wages21_df
```

```python
# Cleaning DataFrame by dropping unnecessary columns df.drop(columns=['B', 'C'], axis=1) axis=1 drops the whole column Stackoverflow

wages21_df = wages21_df.drop(columns=['AREA', 'AREA_TYPE', 'PRIM_STATE', 'NAICS', 'NAICS_TITLE', 'I_GROUP', 'OWN_CODE', 'OCC_CODE', 'H_PCT75', 'H_PCT90', 'A_PCT10', 'A_PCT25', 'ANNUAL', 'HOURLY'], axis=1)
```

```python
wages21_df
```

```python
wages21_df = wages21_df.drop(columns=['O_GROUP', 'EMP_PRSE', 'JOBS_1000', 'LOC_QUOTIENT', 'PCT_TOTAL', 'PCT_RPT', 'MEAN_PRSE', 'H_PCT10', 'H_PCT25', 'A_PCT75', 'A_PCT90'], axis=1)
wages21_df
```

```python
wages21_df
```

```python
wages21_df.info()
```

```python
# Getting rid of duplicates, Pandas documentation https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop_duplicates.html

wages21_df = wages21_df.drop_duplicates()
```

```python
# Resetting index after dropping duplicates. https://stackoverflow.com/questions/28885073/reindexing-after-pandas-drop-duplicates
wages21_df = wages21_df.reset_index(drop=True)
```

```python
wages21_df
```

```python
# 
wages21_df = wages21_df.dropna(axis=0, how='any')
wages21_df.head()
```

```python
# Rename the headers
wages21_df = wages21_df.rename(columns={"AREA_TITLE" : "State",
"OCC_TITLE": "Occupation",
"TOT_EMP": "Total Employees",
"H_MEAN": "Hourly Mean",
"A_MEAN": "Annual Mean",
"H_MEDIAN": "Hourly Median",
"A_MEDIAN": "Annual Median",})

wages21_df.head()
```

```python
# Replacing special characters with nan and then dropping them.
wages21_df = (wages21_df
 .replace('#', np.nan)
 .replace('*', np.nan)
 .replace('**', np.nan)
              
 .dropna()
)

wages21_df.head()
```

```python
# Resetting the index after dropping rows.
wages21_df = wages21_df.reset_index(drop=True)
```

```python
# Renaming the columns  
wages21_df.astype({
    'Hourly Mean':float,
    'Hourly Median':float,
    'Total Employees':float,
    'Annual Mean':float,
    'Annual Median':float,
}).dtypes
```

```python
wages21_df
```

```python
# Exporting Data Frame to csv 
wages21_df.to_csv("../Clean Data/2021wages.csv", index = False)
```
