```python
file_path = "US CPI.csv"
df = pd.read_csv(file_path)
df.tail(50)
```

```python
df.dtypes
```

```python
#url: https://stackoverflow.com/questions/55585851/splitting-a-pandas-object
df['Month'] = df['Yearmon'].str.split('-').str[1]
df['Year'] = df['Yearmon'].str.split('-').str[2]
df
```

```python
#url: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop.html
df = df.drop(columns='Yearmon')
```

```python
df.columns.tolist()
```

```python
df = df[['Month', 'Year', 'CPI']]
df
```

```python
df[['Month', 'Year']] = df[['Month', 'Year']].astype(int)
df.dtypes
```
